# 조회수와 좋아요/싫어요

```python
KRlike=KR[['views','likes','dislikes']]
KRlike.corr()
```

```python
JPlike=JP[['views','likes','dislikes']]
JPlike.corr()
```

한국에 비해 일본의 경우, dislikes와 views 간의 강한 정적인 상관관계를 보임.

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2, 2, 1)
ax = sns.regplot(x='likes', y='views', data=KRlike) 
ax.set_title('KR likes')

bx = fig.add_subplot(2,2,2)
bx = sns.regplot(x='dislikes', y='views', data=KRlike) 
bx.set_title('KR dislikes')


cx = fig.add_subplot(2,2,3)
cx = sns.regplot(x='likes', y='views', data=JPlike) 
ticks = cx.set_xticks([0, 2000000, 4000000])
cx.set_title('JP likes')


dx = fig.add_subplot(2,2,4) 
dx = sns.regplot(x='dislikes', y='views', data=JPlike) 
dx.set_title('JP dislikes')

fig.tight_layout()
```

- 한국의 경우, 
- 좋아요 수와 조회수 간의 정적인 상관관계를 보임.
- 싫어요 수와 조회수 간의 정적인 상관관계를 보이나, 아웃라이어로 인한 과소 추정이 의심됨.
---
- 일본의 경우,
- 좋아요 수와 조회수 간의 정적인 상관관계를 보임.
- 싫어요 수와 조회수 간의 한국보다 더 강력한 정적인 상관관계를 보임.

```python
KRvideos.sort_values(by = ['dislikes'], ascending = [False]).head(4)
```

KR데이터의 아웃라이어 확인 결과, 싫어요 수가 압도적으로 높은 영상은 youtube 자체의 연년 영상임(=조회수가 높을 수밖에 없는 영상)

```python
dislikes = KRvideos['dislikes']
KRvideos_new = KRvideos[dislikes < 800000]
subset_KR_new = KRvideos_new[['views', 'dislikes']]
subset_KR_new.corr()
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(2, 2, 1)
ax = sns.regplot(x='likes', y='views', data=KRlike) 
ax.set_title('KR likes')

bx = fig.add_subplot(2,2,2)
bx = sns.regplot(x='dislikes', y='views', data=subset_KR_new) 
bx.set_title('KR dislikes')


cx = fig.add_subplot(2,2,3)
cx = sns.regplot(x='likes', y='views', data=JPlike) 
ticks = cx.set_xticks([0, 2000000, 4000000])
cx.set_title('JP likes')


dx = fig.add_subplot(2,2,4) 
dx = sns.regplot(x='dislikes', y='views', data=JPlike) 
dx.set_title('JP dislikes')

fig.tight_layout()
```

좋아요 수 싫어요 수 전부 조회수에 비슷한 영향을 미치고 있음
